竟然一次就 AC 了，虽然耗时不短（判断是否可放置需要优化）


```python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        base_res = ["."*n for i in range(n)]
        
        results = []
        combine = []
        def backtrack(idx):
            if idx >= n:
                results.append(combine[:])
            else:
                _line = ["." for _ in range(n)]
                for i in range(n):
                    # 判断该位置是否可以放置
                    valid = True
                    for j in range(1, idx+1):
                        if combine[idx-j][i] == "Q" or (i-j >= 0 and combine[idx-j][i-j] == "Q") or (i+j < n and combine[idx-j][i+j] == "Q"):
                            valid = False
                            break
                    # 不可以放置，直接进入下一层循环
                    if not valid:
                        continue
                    
                    # 可以放置，则进行回溯
                    _line[i] = "Q"
                    combine.append("".join(_line))
                    backtrack(idx + 1)
                    combine.pop()
                    _line[i] = "."
                    
        backtrack(0)
        return results
```

```python
n = 3
print(Solution().solveNQueens(n))
```
