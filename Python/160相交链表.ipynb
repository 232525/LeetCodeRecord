```python
import typing
from typing import List
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

# 方法一：试图构造环形链表，如果两个链表存在相交则能成环，然后基于 LeetCode 142 查找环状链表的入口节点
# Note：在返回时，需要将构造的环取消，还原链表原本结构（题目要求不能对链表进行修改）
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        # 构造一个环状链表
        ptr = headA
        while ptr.next is not None:
            ptr = ptr.next
        ptr.next = headB

        # 查找环状链表的入口节点
        slow, fast = headA, headA
        while fast is not None:
            slow = slow.next
            if fast.next is None:
                # 链表结构还原
                ptr.next = None
                return None
            fast = fast.next.next

            if slow == fast:
                fast = headA
                while fast != slow:
                    fast = fast.next
                    slow = slow.next
                # 链表结构还原
                ptr.next = None
                return slow

        # 链表结构还原
        ptr.next = None
        return None
    
    
# 方法二：使用两个栈，将两个链表的节点入栈，两栈顶元素则为重叠的节点，
# 然后出栈到无重叠节点，则为第一个相交的节点
# 时间复杂度：O(m+n) 空间复杂度：O(m+n)
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        stackA = []
        stackB = []
        ptr = headA
        while ptr is not None:
            stackA.append(ptr)
            ptr = ptr.next
        
        ptr = headB
        while ptr is not None:
            stackB.append(ptr)
            ptr = ptr.next

        while len(stackA) > 0 and len(stackB) > 0:
            ptr1 = stackA.pop()
            ptr2 = stackB.pop()
            if ptr1 != ptr2:
                return ptr1.next

        return headA if len(stackA) == 0 else headB
    
# 方法三：双指针，先获取两条链表的长度
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        nA, nB = 0, 0
        ptr = headA
        while ptr is not None:
            nA += 1
            ptr = ptr.next

        ptr = headB
        while ptr is not None:
            nB += 1
            ptr = ptr.next

        longer = headA if nA > nB else headB
        shorter = headB if nA > nB else headA
        for i in range(abs(nA - nB)):
            longer = longer.next

        while longer != shorter:
            longer = longer.next
            shorter = shorter.next
        
        return shorter
```
