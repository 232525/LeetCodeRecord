```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        def getKthElement(k: int) -> int:
            len1, len2 = len(nums1), len(nums2)
            idx1, idx2 = 0, 0  # 用于舍弃无用值
            while(True):
                # 特殊情况，用于退出循环
                if idx1 == len1:
                    # 数组1全部排除，则中位数位于数组2中
                    return nums2[idx2 + k - 1]
                if idx2 == len2:
                    return nums1[idx1 + k - 1]
                if k == 1:
                    return min(nums1[idx1], nums2[idx2])
                
                
                # 一般情况，下标不能越界
                new_idx1 = min(idx1 + k // 2 - 1, len1 - 1)
                new_idx2 = min(idx2 + k // 2 - 1, len2 - 1)
                """
                if nums1[new_idx1] <= nums2[new_idx2]:
                    # k值更新
                    k = k - (new_idx1 - idx1 + 1)
                    # 更新idx1
                    idx1 = new_idx1 + 1
                else:
                    # k值更新
                    k = k - (new_idx2 - idx2 + 1)
                    # 更新idx2
                    idx2 = new_idx2 + 1
                """
                if nums1[new_idx1] < nums2[new_idx2]:
                    # k值更新
                    k = k - (new_idx1 - idx1 + 1)
                    # 更新idx1
                    idx1 = new_idx1 + 1
                elif nums1[new_idx1] > nums2[new_idx2]:
                    # k值更新
                    k = k - (new_idx2 - idx2 + 1)
                    # 更新idx2
                    idx2 = new_idx2 + 1
                else:
                    k = k - (new_idx1 - idx1 + 1) - (new_idx2 - idx2 + 1)
                    if k == 0:
                        return nums1[new_idx1]
                    else:
                        idx1 = new_idx1 + 1
                        idx2 = new_idx2 + 1
        
        len1, len2 = len(nums1), len(nums2)
        len_sum = len1 + len2
        k = (len_sum - 1) // 2 + 1
        if len_sum % 2 != 0:
            # 长度和为奇数，中位数只有一个
            return getKthElement(k)
        else:
            # 长度和为偶数，中位数为两个的均值
            return (getKthElement(k) + getKthElement(k+1)) / 2.
```
