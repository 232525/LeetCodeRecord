```python
class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        len_word = len(words[0])        # 一个单词的长度
        num_words = len(words)          # 单词列表的长度，即单词个数
        
        # 字符串s的长度
        len_s = len(s)
        # left, right：用来判断子串是否为words中的单词
        left = 0
        right = left + len_word
        # 结果
        results = []
        while left + num_words * len_word <= len_s:
            sub_s = s[left: right]  # 获取子字符串
            if sub_s in words:
                # 如果存在子串sub_s为words中的单词，则进一步判断是否满足题目要求
                if self.matchWords(s[left: left + num_words * len_word], words.copy(), len_word, num_words):
                    results.append(left)
            left += 1
            right += 1
                
        return results
    
    # 函数是否可进一步优化？存在重复比较的情况？
    def matchWords(self, sub_s, words, len_word, num_words):
        if num_words == 1:
            return sub_s == words[0]
        else:
            sub_sub_s = sub_s[:len_word]
            if sub_sub_s not in words:
                return False
            else:
                words.remove(sub_s[:len_word])
                return True and (self.matchWords(sub_s[len_word:], words, len_word, num_words-1))
```

```python
s = "barfoothefoobarman"
words = ["foo","bar"]

s = "wordgoodgoodgoodbestword"
words = ["word","good","best","word"]

s = "barfoofoobarthefoobarman"
words = ["bar","foo","the"]
print(Solution().findSubstring(s, words))
```
