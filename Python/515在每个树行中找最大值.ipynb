```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# BFS：为了计算每一层节点的最大值，需要同时知道该节点的层数，因此在队列中同时存入节点及其层数
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        INT_MIN = -2 ** 31

        ans = []
        curr_layer = 0       # 当前层
        
        # BFS
        queue = deque()      # 队列中同时存储当前节点及节点所在的层数，[node, node's layer]
        if root is not None:
            queue.append([root, 0])
            ans.append(root.val)
        while len(queue) > 0:
            node, layer = queue.popleft()
            if layer == curr_layer + 1:
                ans.append(node.val)          # 已经到了新一层，将当前节点的值加入到 ans 中
                curr_layer = layer
            ans[-1] = max(ans[-1], node.val)  # 计算最大值
            if node.left is not None:
                queue.append([node.left, layer+1]) 
            if node.right is not None:
                queue.append([node.right, layer+1])

        return ans
    
# 剑指offer II 044 题解：使用两个变量，curr_nodes 和 next_nodes 存储每一层的节点个数
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        INT_MIN = -2 ** 31

        ans = []
        curr_nodes = 0
        next_nodes = 0
        
        # BFS
        queue = deque()      # 队列中同时存储当前节点及节点所在的层数，[node, node's layer]
        if root is not None:
            queue.append(root)
            curr_nodes = 1

        curr_max = INT_MIN
        while len(queue) > 0:
            node = queue.popleft()
            curr_nodes -= 1
            curr_max = max(curr_max, node.val)  # 计算最大值

            # BFS 将节点加入 queue 中
            if node.left is not None:
                queue.append(node.left)
                next_nodes += 1
            if node.right is not None:
                queue.append(node.right)
                next_nodes += 1
            
            # 统计新层
            if curr_nodes == 0:
                curr_nodes = next_nodes
                next_nodes = 0
                ans.append(curr_max)
                curr_max = INT_MIN

        return ans
    
# 使用两个队列进行实现，两个队列分别存储不同层的节点
# 逻辑上更为简洁
class Solution:
    def largestValues(self, root: Optional[TreeNode]) -> List[int]:
        INT_MIN = -2 ** 31

        queue1 = deque()
        queue2 = deque()
        if root is not None:
            queue1.append(root)

        ans = []
        curr_max = INT_MIN
        while len(queue1) > 0:
            node = queue1.popleft()
            curr_max = max(curr_max, node.val)

            if node.left is not None:
                queue2.append(node.left)
            if node.right is not None:
                queue2.append(node.right)

            # 当 queue1 为空时，表明当前层的节点已经读取完，
            # 可以将 curr_max 存入 ans，
            # 并将 queue2 赋值给 queue1 作为新的 queue1，并新建一个空队列作为 新的 queue2
            if len(queue1) == 0:
                ans.append(curr_max)
                curr_max = INT_MIN
                queue1 = queue2
                queue2 = deque()

        return ans
```
