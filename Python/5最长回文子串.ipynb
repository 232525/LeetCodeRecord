### 最长回文子序列
__方法一：动态规划（时间复杂度O(n^2)）__

令P(i,j)表示字符串s的子串s[i:j]是否为回文串，状态转移方程如下：

P(i, j) = P(i+1, j-1) & s[i] == s[j]

```python
class Solution1:
    def longestPalindrome(self, s: str) -> str:
        len_s = len(s)
        if len_s < 2:
            return s
        
        # 最长回文长度，至少为1
        max_len = 1
        begin = 0
        # 状态矩阵
        dp = [[False] * len_s for _ in range(len_s)]
        # 单个字符自然是回文
        for i in range(len_s):
            dp[i][i] = True
        
        # 递推求解
        # 子串长度，必须从 短串 --> 长串 循环，
        # 因为长串是否为回文串依赖与短串的状态
        for L in range(2, len_s+1):
            # i-左边界；j-右边界
            for i in range(len_s):
                j = L + i - 1
                # 越界直接退出循环
                if j >= len_s:
                    break
                    
                if s[i] != s[j]:
                    dp[i][j] = False
                else:
                    if j - i < 3:
                        dp[i][j] = True
                    else:
                        dp[i][j] = dp[i+1][j-1]
                
                # 记录回文起始位置及长度
                if dp[i][j] and j - i + 1 > max_len:
                    max_len = j - i + 1
                    begin = i
        
        return s[begin : begin + max_len]
    
# 遍历中心点，向两边扩展
class Solution2:    
    def longestPalindrome(self, s: str) -> str:
        result = ""
        len_s = len(s)
        # 字符串长度 < 2
        if len_s < 2:
            return s
        
        # 字符串长度 >= 2
        # 记录结果
        begin = 0
        len_result = 1
        
        # i -> 中心点遍历
        for i in range(len_s):
            # 奇数长度扩展
            delta_odd = min(i, len_s-1-i)
            left, right = i, i
            for j in range(1, delta_odd + 1):
                left = i - j
                right = i + j
                if s[left] != s[right]:
                    left = left + 1
                    right = right - 1
                    break
            if right - left + 1 > len_result:
                len_result = right - left + 1
                begin = left
            
            # 偶数长度扩展
            # 越界，或者两个字符不相同，直接跳过
            if i+1 >= len_s or s[i] != s[i+1]:
                continue
            delta_even = min(i, len_s-1-i-1)
            left, right = i, i+1
            for j in range(1, delta_even + 1):
                left = i - j
                right = i + 1 + j
                if s[left] != s[right]:
                    left = left + 1
                    right = right - 1
                    break
                    
            if right - left + 1 > len_result:
                len_result = right - left + 1
                begin = left
                        
        print('result:', begin, len_result)
        return s[begin: begin+len_result]

# 分为两个函数
class Solution:
    def expandAroundCenter(self, s: str, left: int, right: int) -> [int, int]:
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left  -= 1
            right += 1
        # 不满足条件时，返回值right < left
        return left + 1, right - 1
        
    def longestPalindrome(self, s: str) -> str:
        len_s = len(s)
        # 字符串长度 < 2，直接输出
        if len_s < 2:
            return s
        
        # 字符串长度 >= 2，进行扩展判断
        # 记录结果
        begin = 0
        len_result = 1
        
        # i -> 中心点遍历
        for i in range(len_s):
            # 奇数长度扩展
            left, right = self.expandAroundCenter(s, i, i)
            if len_result < (right - left + 1):
                len_result = right - left + 1
                begin = left
            
            # 偶数长度扩展
            left, right = self.expandAroundCenter(s, i, i + 1)
            if len_result < (right - left + 1):
                len_result = right - left + 1
                begin = left
                        
        # print('result:', begin, len_result)
        return s[begin: begin+len_result]
    
class Solution_:
    def expandAroundCenter(self, s, left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return left + 1, right - 1

    def longestPalindrome(self, s: str) -> str:
        start, end = 0, 0
        for i in range(len(s)):
            left1, right1 = self.expandAroundCenter(s, i, i)
            left2, right2 = self.expandAroundCenter(s, i, i + 1)
            if right1 - left1 > end - start:
                start, end = left1, right1
            if right2 - left2 > end - start:
                start, end = left2, right2
        return s[start: end + 1]
```
