```python
# 用 dict 作为哈希表
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        hash_s1 = {}
        if len(s2) < len(s1):
            return False
            
        # 统计 s1 字符信息
        for char in s1:
            if char not in hash_s1:
                hash_s1[char] = 1
            else:
                hash_s1[char] += 1
        
        hash_sub_s2 = {}
        for i in range(len(s1)):
            char = s2[i]
            if char not in hash_sub_s2:
                hash_sub_s2[char] = 1
            else:
                hash_sub_s2[char] += 1

        left, right = 0, len(s1) - 1
        while right < len(s2):
            _ = self.check(hash_sub_s2, hash_s1)
            if _:
                return True
            else:
                if right + 1 >= len(s2):
                    break
                hash_sub_s2[s2[left]] -= 1
                if s2[right + 1] not in hash_sub_s2:
                    hash_sub_s2[s2[right + 1]] = 1
                else:
                    hash_sub_s2[s2[right + 1]] += 1
                left += 1
                right += 1

        return False

    # 判断两个 哈希表 信息是否一致（即包含的字符统计是否一致）
    def check(self, hash_sub_s2, hash_s1):
        for char, cnt in hash_sub_s2.items():
            if char not in hash_s1 and cnt > 0:
                return False
            if char in hash_s1 and cnt != hash_s1[char]:
                return False
        return True
    
# 使用 List 作为哈希表
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s2) < len(s1):
            return False
            
        hash_s1 = [0] * 26
        hash_sub_s2 = [0] * 26
        
        # 统计 s1 字符信息
        for i, char in enumerate(s1):
            hash_s1[ord(char) - ord('a')] += 1
            hash_sub_s2[ord(s2[i]) - ord('a')] += 1
        
        if self.check(hash_sub_s2, hash_s1):
            return True

        right = len(s1)
        while right < len(s2):
            hash_sub_s2[ord(s2[right - len(s1)]) - ord('a')] -= 1
            hash_sub_s2[ord(s2[right]) - ord('a')] += 1
            _ = self.check(hash_sub_s2, hash_s1)
            if _:
                return True
            else:
                right += 1

        return False

    # 判断两个 哈希表 信息是否一致（即包含的字符统计是否一致）
    def check(self, hash_sub_s2, hash_s1):
        for i in range(26):
            if hash_s1[i] - hash_sub_s2[i] != 0:
                return False
        return True
    
# 使用 List 作为哈希表 -- 空间优化，只使用一个 List
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        n1, n2 = len(s1), len(s2)
        if n2 < n1:
            return False
            
        hash_infos = [0] * 26
        
        # 统计 s1 字符信息
        for i in range(n1):
            hash_infos[ord(s1[i]) - ord('a')] += 1
            hash_infos[ord(s2[i]) - ord('a')] -= 1
        
        # 判断第一个子串
        if self.check(hash_infos):
            return True

        # 判断剩余子串
        for i in range(n1, n2):
            hash_infos[ord(s2[i - n1]) - ord('a')] += 1
            hash_infos[ord(s2[i]) - ord('a')] -= 1
            _ = self.check(hash_infos)
            if _:
                return True

        return False

    # 判断两个 哈希表 信息是否一致（即包含的字符统计是否一致）
    def check(self, hash_infos):
        for i in range(26):
            if hash_infos[i] != 0:
                return False
        return True
```
